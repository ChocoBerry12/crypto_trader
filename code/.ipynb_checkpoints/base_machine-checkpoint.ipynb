```python
from abc import ABC, abstractmethod

class Machine(ABC):
    
    @abstractmethod
    def get_sign(self):
        """암호화된 sign 반환 메서드.
            US 거래소에서 주로 쓰임.
        """
        pass
    
    @abstractmethod
    def get_nonce(self):
        """nonce 반환 메서드.
        """
        pass
    
    @abstractmethod
    def get_timestamp(self):
        """ timestamp 반환하는 메서드.
        """
        pass
    
    @abstractmethod
    def get_user_info(self):
        """현재 사용자 정보를 받기 위한 메소드입니다.
        """
        pass
    
    @abstractmethod
    def get_token(self):
        """액세스토큰 정보를 받기 위한 메소드입니다.
            토큰은 주로 KR 거래소에서 주로 쓰임.
        """
        pass

    @abstractmethod
    def set_token(self):
        """액세스토큰 정보를 만들기 위한 메소드입니다.
        """
        pass
    
    @abstractmethod
    def get_balances(self):
        """사용자의 지갑정보를 조회하는 메소드입니다.
        """
        pass
    @abstractmethod
    def buy_order(self):
        """매수 주문을 실행하는 메소드입니다. 
        """
        pass

    @abstractmethod
    def sell_order(self):
        """매도 주문을 실행하는 메소드입니다.
        """
        pass

    @abstractmethod
    def get_closed_orders(self):
        """체결정보를 얻어오기 위한 메소드입니다.
        """
        pass
    
    @abstractmethod
    def get_open_orders(self):
        """미체결 주문 정보 반환
        """
        pass
    
    @abstractmethod
    def cancel_order(self):
        """취소 주문을 실행하는 메소드입니다.
        """
        pass
    
    @abstractmethod
    def move_order(self):
        """주문을 수정하는 메서드
        """
        pass
    
    @abstractmethod
    def get_order_status(self):
        """사용자의 주문정보 상세정보를 조회하는 메소드입니다. 
        """
        pass
    
    @abstractmethod
    def get_ticker(self):
        """마지막 체결정보(Tick)을 얻기 위한 메소드입니다.
        """
        pass
    
    @abstractmethod
    def get_history_deposit_withdraw(self):
        """입출금 내역 반환
        """
        pass    
```
